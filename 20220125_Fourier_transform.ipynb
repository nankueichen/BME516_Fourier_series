### Periodic function
A function $f(x)$ of one variable $x$ is periodic, with period $T$: $f(x+T) = f(x)$. </br>
It follows that if $T$ is a period of $f$, then $mT$ is also a period for every integer $m$ </br>
For example, the functions
$$ \sin\frac{m\pi x}{L}, \: \cos\frac{m\pi x}{L}, \: \mathrm{e}^{i\frac{m\pi x}{L}}$$ </br>
are periodic with fundamental period $T=\frac{2L}{m}$

___

### Mutually orthogonal set
The functions
$$1,\: \cos\frac{m\pi x }{L}, \: \sin\frac{m\pi x}{L}$$
form a mutually orthogonal set on the interval $[-L,L]$ as well as on every interval $[a,a+2L]$.

$$\int_{-L}^L \cos\frac{m\pi x}{L} \cos\frac{n\pi x}{L} dx = \left\{%
\begin{array}%
0, & m \neq n, \\
L, & m=n,
\end{array}%
\right. %
$$

$$\int_{-L}^L \cos\frac{m\pi x}{L} \sin\frac{n\pi x}{L} dx = 0 $$

$$\int_{-L}^L \sin\frac{m\pi x}{L} \sin\frac{n\pi x}{L} dx = \left\{%
\begin{array}%
0, & m \neq n, \\
L, & m=n,
\end{array}%
\right. %
$$
and
$$\int_{-L}^L \sin \frac{m\pi x}{L} = \int_{-L}^L \cos \frac{m\pi x}{L} = 0$$
__Proof:__ </br>
First, with the equality $$ \sin\alpha \sin\beta = \frac{1}{2}(\cos(\alpha-\beta)-cos(\alpha+\beta))$$, for $m \neq n$ we have the following:
$$ 
\begin{align*}
\int_{-L}^L \sin\frac{m\pi x}{L} \sin\frac{n\pi x}{L} dx =
& \frac{1}{2} \int_{-L}^L \cos\frac{(m-n)\pi x}{L}dx - \frac{1}{2} \int_{-L}^L \cos\frac{(m+n)\pi x}{L}dx \\
& \left. =\frac{1}{2}\Bigg(\frac{\sin\frac{(m-n)\pi x}{L}}{\frac{(m-n)\pi}{L}}\Bigg)\right|_{-L}^L -\left. \frac{1}{2}\Bigg(\frac{\sin\frac{(m+n)\pi x}{L}}{\frac{(m+n)\pi}{L}}\Bigg)\right|_{-L}^L \\
& =0
\end{align*}
$$

### Fourier Series
This series
$$f(x) = \frac{a_0}{2} + \sum_{m=1}^\infty \big( a_m \cos \frac{m\pi x}{L} + b_m \sin \frac{m\pi x}{L}\big) $$
consists of $2L$-periodic functions. </br></br>
Let us carry out the following integration, term by term, for each $n=1,2, ... $
$$ 
\begin{align*}
\int_{-L}^L f(x) \cos\frac{n\pi x}{L} dx =& \frac{a_0}{2}\int_{-L}^L\cos\frac{n\pi x}{L} dx 
+ \sum_{m=1}^\infty a_m \int_{-L}^L\cos\frac{m\pi x}{L}\cos\frac{n\pi x}{L} dx \\
& + \sum_{m=1}^\infty b_m \int_{-L}^L\sin\frac{m\pi x}{L}\sin\frac{n\pi x}{L} dx  \\
\end{align*}
$$
Based on orthogonality relations, we now have </br>
$$a_n = \frac{1}{L}\int_{-L}^L f(x) \cos \frac{n\pi x}{L} dx, \;\;\; n=1,2,...$$
__Exercise__: </br>
Show that $$a_0 = \frac{1}{L}\int_{-L}^L f(x) dx, \;\;\; n=1,2,...$$
__Exercise__: </br>
Show that
$$b_n = \frac{1}{L}\int_{-L}^L f(x) \sin \frac{n\pi x}{L} dx, \;\;\; n=1,2,...$$

Note that, when $L=\pi$, the Fourier series transforms to
$$f(x) = \frac{a_0}{2} + \sum_{m=1}^\infty \big( a_m \cos mx + b_m \sin mx\big) $$
</br>
__Homework__: </br>
Using [Euler's fomula](https://en.wikipedia.org/wiki/Euler%27s_formula_), the above listed Fourier series could be written in the complex form
$$f(x) = \sum_{n=-\infty}^\infty c_n \mathrm{e}^{inx}$$
where the coefficients $c_n = c_n(f)$ are </br>
$$c_n= \left\{%
\begin{array}%
\frac{a_n}{2}+\frac{b_n}{2i}, & n=1,2,..., \\
\frac{a_0}{2}, & n=0, \\
\frac{a_{-n}}{2}-\frac{b_{-n}}{2i}, & n=-1,-2,...,
\end{array}%
\right. %
$$
$c_n(f)$ is the $n$th __Fourier coefficient__ of $f$, with
$$c_n(f) = \frac{1}{2\pi}\int_{-\pi}^{\pi}f(x) \mathrm{e}^{-inx}dx$$

### Fourier series

```julia
L = 500;
N = 2*L;
timeArray = Array(1:N);
Array = [0, 3, 8];
ampCosArray = [1.5, 3.2, 0];
ampSinArray = [0, 0, 0.8];

Array1 = ampCosArray[1].*cos.(2**Array[1].*timeArray/N) + ampSinArray[1].*sin.(2**Array[1].*timeArray/N);
Array2 = ampCosArray[2].*cos.(2**Array[2].*timeArray/N) + ampSinArray[2].*sin.(2**Array[2].*timeArray/N);;
Array3 = ampCosArray[3].*cos.(2**Array[3].*timeArray/N) + ampSinArray[3].*sin.(2**Array[3].*timeArray/N);;
PyPlot.figure(1,figsize=(8,2));PyPlot.plot(timeArray, Array1,"g",timeArray, Array2,"k",timeArray, Array3,"b");
```

```julia
newArray = Array1 + Array2 + Array3;
PyPlot.figure(1,figsize=(8,2));PyPlot.plot(timeArray,newArray,"b");
PyPlot.figure(2,figsize=(8,2));PyPlot.plot([newArray;newArray;newArray;newArray],"b");
```

```julia
@show Array;
@show ampCosArray
@show ampSinArray;
```

with $a_0=1.5\times2 = 3$, $a_3=3.2$, $b_8=0.8$

Using Euler's formula, the Fourier series could be written in the complex form
$$f(x) = \sum_{n=-\infty}^\infty c_n \mathrm{e}^{inx}$$
$c_n(f)$ is the $n$th __Fourier coefficient__ of $f$:
$$c_n(f) = \frac{1}{2\pi}\int_{-\pi}^{\pi}f(x) \mathrm{e}^{-inx}dx$$

___

$$f(x) = \sum_{n=-\infty}^\infty c_n \mathrm{e}^{i\frac{n\pi x}{L}}$$
where $c_n(f)$ is the $n$th __Fourier coefficient__ of $f$:
$$c_n(f) = \frac{1}{2L}\int_{-L}^{L}f(x) \mathrm{e}^{-i\frac{n\pi x}{L}}dx$$

```julia
L = 500;
N = 2L;
timeArray = Array(1:N);
Array = [0, 3, 8];
ampArray = [complex(1.5,0), complex(2.2,0), complex(0,0.56)];

Array1 = ampArray[1].*exp.(-i*2**Array[1].*timeArray/N) 
Array2 = ampArray[2].*exp.(-i*2**Array[2].*timeArray/N) 
Array3 = ampArray[3].*exp.(-i*2**Array[3].*timeArray/N) 
PyPlot.figure(1,figsize=(8,2));PyPlot.plot(timeArray, real.(Array1),"g",timeArray, real.(Array2),"k",timeArray, real.(Array3),"b");
```

```julia
newArray = Array1 + Array2 + Array3;
PyPlot.figure(1,figsize=(8,2));PyPlot.plot(timeArray,real.(newArray),"b");
PyPlot.figure(2,figsize=(8,2));PyPlot.plot([real.(newArray);real.(newArray);real.(newArray);real.(newArray)],"b");
```

```julia
@show Array
@show ampArray;

@show abs.(ampArray);
@show angle.(ampArray);

@show real.(ampArray);
@show imag.(ampArray);
```

```julia
@show transpose(exp.(i*2**Array[1].*timeArray/N)) * (newArray) / N 
@show transpose(exp.(i*2**Array[2].*timeArray/N)) * (newArray) / N 
@show transpose(exp.(i*2**Array[3].*timeArray/N)) * (newArray) / N

@show transpose(exp.(i*2** 23 .*timeArray/N)) * (newArray) / N ;
```
